```python
data_since = pd.read_csv("/kaggle/input/data-science-job-salaries/ds_salaries.csv")
df = data_since.copy()
df = df.dropna()
df= df.drop('Unnamed: 0',axis=1)
dms = pd.get_dummies(df[['experience_level', 'employment_type', 'job_title','salary_currency','employee_residence','company_location','company_size']])
y = df["salary"]
X_ = df.drop(['salary','experience_level', 'employment_type', 'job_title','salary_currency','employee_residence','company_location','company_size'], axis = 1).astype("int64")
X = pd.concat([X_, dms[:]], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.25, 
                                                    random_state=42)
```

```python
lambdalar = 10**np.linspace(10,-2,100)*0.5 

ridge_model = Ridge()
katsayilar = []

for i in lambdalar:
    ridge_model.set_params(alpha = i)
    ridge_model.fit(X_train, y_train) 
    katsayilar.append(ridge_model.coef_) 
    

    
ax = plt.gca()
ax.plot(lambdalar, katsayilar) 
ax.set_xscale('log') 

plt.xlabel('Lambda(Alpha) Değerleri')
plt.ylabel('Katsayılar/Ağırlıklar')
plt.title('Düzenlileştirmenin Bir Fonksiyonu Olarak Ridge Katsayıları');
```

```python
# Model
data_since = pd.read_csv("/kaggle/input/data-science-job-salaries/ds_salaries.csv")
df = data_since.copy()
df = df.dropna()
df= df.drop('Unnamed: 0',axis=1)
dms = pd.get_dummies(df[['experience_level', 'employment_type', 'job_title','salary_currency','employee_residence','company_location','company_size']]) 
y = df["salary"]
X_ = df.drop(['salary','experience_level', 'employment_type', 'job_title','salary_currency','employee_residence','company_location','company_size'], axis = 1).astype("int64") 
X = pd.concat([X_, dms[:]], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso_model = Lasso(alpha = 0.1).fit(X_train, y_train)
lasso_model
```

```python
lasso_model.coef_
```

```python
lasso = Lasso()
lambdalar = 10**np.linspace(10,-2,100)*0.5 
katsayilar = []

for i in lambdalar:
    lasso.set_params(alpha=i)
    lasso.fit(X_train, y_train)
    katsayilar.append(lasso.coef_)
    
ax = plt.gca()
ax.plot(lambdalar*2, katsayilar)
ax.set_xscale('log')
plt.axis('tight')
plt.xlabel('alpha')
plt.ylabel('weights')
```

```python

# Model
data_since = pd.read_csv("/kaggle/input/data-science-job-salaries/ds_salaries.csv")
df = data_since.copy()
df = df.dropna()
df= df.drop('Unnamed: 0',axis=1)
dms = pd.get_dummies(df[['experience_level', 'employment_type', 'job_title','salary_currency','employee_residence','company_location','company_size']]) 
y = df["salary"]
X_ = df.drop(['salary','experience_level', 'employment_type', 'job_title','salary_currency','employee_residence','company_location','company_size'], axis = 1).astype("int64") 
X = pd.concat([X_, dms[:]], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
enet_model = ElasticNet().fit(X_train, y_train)
```

```python
enet_model.coef_
```

```python
enet_model.intercept_
```

```python
# * Tahmin
enet_model
```

```python
enet_model.predict(X_test)
```

```python
y_pred = enet_model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
enet_cv_model = ElasticNetCV(cv = 10, random_state = 0).fit(X_train, y_train)
```
